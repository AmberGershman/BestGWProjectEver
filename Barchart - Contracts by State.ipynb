```python
import os
import csv
import pandas as pd
import matplotlib.pyplot as plt
```

```python
procurement_data = "Federal_Overview.csv"
procurement_data_df = pd.read_csv(procurement_data)
procurement_data_df.head()
```

```python
procurement_data_bystate_df = procurement_data_df.groupby(["State"])['Dollars Obligated'].sum()
procurement_data_bystate_df 
# procurement_data_bystate_df_sort = procurement_data_bystate_df.sort_values(by='Dollars Obligated', ascending=True)

procurement_data_bystate_df.plot.barh()

plt.title("Dollars Obligated by State")
plt.ylabel("State")
plt.xlabel("Dollars Obligated")
plt.grid(True)

plt.savefig("dollarsoblig_by_state.png")
```
